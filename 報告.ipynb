```
import pandas as pd
d = pd.read_excel ("456.xlsx")  #d為檔案資料
d.head(5)
```

```
d_107 = d[d["year"] == 107] #d_107為107年的資料
d_107.shape
```

```
d_107
```

```
d_98 = d[d["year"] == 98] #d_98為98年的資料
d_98
```

```
d.pivot_table(
    values=['Government bond'], 
    index='country', 
    columns=['year'])   #d.pivot_table為政府公債的資料
```

```
d_107_bond = d_107["Government bond"] #d_107_bond為107年政府公債的資料
d_107_bond.values
```

```
country_name = ["Taipei", "New taipei", "Taoyuan", "Taichung", "Tainan", "Kaohsiung", "Miaoli"] #country_name為七個城市名
```

```
import matplotlib.pyplot as plt
country_bond_107 = len(d_107_bond.values)  #country_bond_107為107年政府公債資料的長度
plt.bar(range(country_bond_107), d_107_bond.values) #直方圖
plt.xticks(range(country_bond_107), country_name)
plt.title("How many bond in each county_107")
plt.xlabel("country")
plt.ylabel("government bond")
plt.show()
```

```
d_98_bond = d_98["Government bond"] #d_98_bond為98年政府公債的資料
d_98_bond.values
```

```
import matplotlib.pyplot as plt
country_bond_98 = len(d_98_bond.values) #country_bond_98為98年政府公債資料的長度
plt.bar(range(country_bond_98), d_98_bond.values)  #直方圖
plt.xticks(range(country_bond_98), country_name)
plt.title("How many bond in each county_98")
plt.xlabel("country")
plt.ylabel("government bond")
plt.show()
```

```
Taipei = d[d["country"] == "Taipei"]  #Taipei為資料裡的country是台北
Miaoli = d[d["country"] == "Miaoli"]  #Miaoli為資料裡的country是苗栗
plt.plot(Taipei["year"], Taipei["Government bond"], label = "Taipei")  #折線圖#台北在98-107年的政府公債
plt.plot(Miaoli["year"], Miaoli["Government bond"], label = "Miaoli")  #折線圖#苗栗在98-107年的政府公債
plt.legend()
plt.show()
```

```
Taipei = d[d["country"] == "Taipei"]  #Taipei為資料裡的country是台北
Newtaipei = d[d["country"] == "New taipei"]  #Newtaipei為資料裡的country是新北
plt.plot(Taipei["year"], Taipei["Government bond"], label = "Taipei")  #折線圖#台北在98-107年的政府公債
plt.plot(Newtaipei["year"], Newtaipei["Government bond"], label = "New taipei")  #折線圖#新北在98-107年的政府公債
plt.legend()
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]  #subset_df為資料裡的country
  plt.plot(subset_df["year"], subset_df["Government bond"], label = country) #折線圖#政府公債
plt.legend()
plt.xlabel("Year")
plt.ylabel("Government bond")
plt.title("country's government bond in 98-107")
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]
  plt.plot(subset_df["year"], subset_df["Government tax"], label = country) #折線圖#政府稅收
plt.legend()
plt.xlabel("Year")
plt.ylabel("Government tax")
plt.title("country's government tax in 98-107")
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]
  plt.plot(subset_df["year"], subset_df["Government spending"], label = country) #折線圖#政府支出
plt.legend()
plt.xlabel("Year")
plt.ylabel("Government spending")
plt.title("country's government spending in 98-107")
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]
  plt.plot(subset_df["year"], subset_df["Central subsidy"], label = country) #折線圖#中央補助
plt.legend()
plt.xlabel("Year")
plt.ylabel("Central subsidy")
plt.title("country's central subsidy in 98-107")
plt.show()
```
